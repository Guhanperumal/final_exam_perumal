## Data cleaning

### Subtask:
Address missing values and outliers based on the findings from the inspection and analysis steps. Add comments to the code explaining the cleaning steps.


**Reasoning**:
Address missing values and outliers in the dataset based on the previous analysis.



```python
# 1. Address missing values.
# Based on the isnull().sum() output, some columns have a very high percentage of missing values
# (e.g., 'Closed Won Reason', 'Annual contract value', 'Medium of last booking in meetings tool').
# Dropping columns with more than 50% missing values is a reasonable strategy to avoid introducing
# bias through imputation when most data is missing.

# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().sum() / len(df) * 100

# Identify columns to drop (e.g., more than 50% missing)
cols_to_drop = missing_percentage[missing_percentage > 50].index

# Drop the identified columns from the DataFrame
df_cleaned = df.drop(columns=cols_to_drop)
print(f"Dropped columns with more than 50% missing values: {list(cols_to_drop)}")

# For remaining columns with missing values, we can consider imputation.
# For numerical columns, impute with the median as it's less sensitive to outliers.
# For categorical columns, impute with the mode (most frequent value).

# Identify numerical and categorical columns in the cleaned DataFrame
numerical_cols_cleaned = df_cleaned.select_dtypes(include=['number']).columns
categorical_cols_cleaned = df_cleaned.select_dtypes(include=['object']).columns

# Impute missing values in numerical columns with the median
for col in numerical_cols_cleaned:
    if df_cleaned[col].isnull().sum() > 0:
        median_val = df_cleaned[col].median()
        df_cleaned[col].fillna(median_val, inplace=True)
        print(f"Imputed missing values in '{col}' with the median ({median_val}).")

# Impute missing values in categorical columns with the mode
for col in categorical_cols_cleaned:
    if df_cleaned[col].isnull().sum() > 0:
        # Calculate mode, handle cases where mode might be empty (e.g., all NaNs or multiple modes)
        mode_val = df_cleaned[col].mode()
        if not mode_val.empty:
            df_cleaned[col].fillna(mode_val[0], inplace=True)
            print(f"Imputed missing values in '{col}' with the mode ({mode_val[0]}).")
        else:
            print(f"Could not impute missing values in '{col}' as mode is not well-defined.")


# Verify that missing values have been addressed
print("\nMissing values after handling:")
display(df_cleaned.isnull().sum())


# 2. Address outliers in numerical columns.
# Based on the box plots, 'Number of times contacted', 'Number of Sales Activities', and 'Amount in company currency'
# show potential outliers.
# A common strategy is capping (winsorizing) outliers at a certain percentile (e.g., 95th or 99th)
# or using the IQR method to define bounds and cap values outside these bounds.
# We will use the IQR method to cap outliers for 'Number of times contacted' and 'Number of Sales Activities'.
# For 'Amount in company currency', given its potential skewness and importance, we might cap at a higher percentile.

# Define columns for outlier handling
outlier_cols = ['Number of times contacted', 'Number of Sales Activities']
high_percentile_outlier_col = 'Amount in company currency'

# Function to cap outliers using IQR
def cap_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap values below the lower bound and above the upper bound
    return series.clip(lower=lower_bound, upper=upper_bound)

# Apply IQR capping to specified columns
for col in outlier_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = cap_outliers_iqr(df_cleaned[col])
        print(f"Capped outliers in '{col}' using the IQR method.")
    else:
        print(f"Column '{col}' not found in cleaned DataFrame.")


# Cap outliers in 'Amount in company currency' at the 99th percentile
if high_percentile_outlier_col in df_cleaned.columns:
    upper_bound_percentile = df_cleaned[high_percentile_outlier_col].quantile(0.99)
    df_cleaned[high_percentile_outlier_col] = df_cleaned[high_percentile_outlier_col].clip(upper=upper_bound_percentile)
    print(f"Capped outliers in '{high_percentile_outlier_col}' at the 99th percentile ({upper_bound_percentile}).")
else:
    print(f"Column '{high_percentile_outlier_col}' not found in cleaned DataFrame.")


# Verify that outliers have been addressed by visualizing distributions again if necessary.
# We can replot the box plots for the columns where outliers were handled.
numerical_cols_after_outlier_handling = outlier_cols + [high_percentile_outlier_col]

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols_after_outlier_handling):
    if col in df_cleaned.columns:
        plt.subplot(1, 3, i + 1)
        sns.boxplot(y=df_cleaned[col])
        plt.title(f'Box Plot of {col} After Outlier Handling')
        plt.ylabel(col)
    else:
        print(f"Could not plot box plot for '{col}' as it's not in the cleaned DataFrame.")

plt.tight_layout()
plt.show()
```

```python
# Consider creating a new categorical feature based on deal duration.
# Define bins for deal duration to create categories (e.g., Short, Medium, Long).
# We can use quantiles or specific day ranges based on the distribution observed from the describe() output.
# Based on the describe output (min=-123, 25%=34, 50%=104, 75%=104, max=231),
# let's define categories based on quantiles:
# < 25th percentile: Short
# 25th percentile to 75th percentile: Medium
# > 75th percentile: Long

quantiles = df_cleaned['Deal Duration (days)'].quantile([0.25, 0.75])
short_threshold = quantiles[0.25]
long_threshold = quantiles[0.75]

# Define the bins and labels for the new categorical feature
bins = [df_cleaned['Deal Duration (days)'].min() - 1, short_threshold, long_threshold, df_cleaned['Deal Duration (days)'].max() + 1] # Adjust bins to include min/max
labels = ['Short', 'Medium', 'Long']

# Create the new categorical feature 'Deal Duration Category'
df_cleaned['Deal Duration Category'] = pd.cut(df_cleaned['Deal Duration (days)'], bins=bins, labels=labels, include_lowest=True)

# Display the distribution of the new category
print("\nDistribution of Deal Duration Category:")
display(df_cleaned['Deal Duration Category'].value_counts())

# Analyze the relationship between 'Deal Duration Category' and 'Deal Stage' using a stacked bar chart.
# First, create a crosstab to count the occurrences of each Deal Stage within each Deal Duration Category.
crosstab_duration_stage = pd.crosstab(df_cleaned['Deal Duration Category'], df_cleaned['Deal Stage'])

# Plot a stacked bar chart
crosstab_duration_stage.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Deal Stage Distribution by Deal Duration Category')
plt.xlabel('Deal Duration Category')
plt.ylabel('Number of Deals')
plt.xticks(rotation=0) # Keep x-labels horizontal
plt.legend(title='Deal Stage', bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot
plt.tight_layout()
plt.show()

# Add comments explaining the stacked bar chart
# This stacked bar chart shows the proportion of each deal stage within each deal duration category.
# It helps understand if the duration of a deal is related to its final stage or progress.
# For example, are "Closed won" deals more likely to be in a specific duration category?
# Do "Closed Lost" deals tend to have a certain duration?
```
